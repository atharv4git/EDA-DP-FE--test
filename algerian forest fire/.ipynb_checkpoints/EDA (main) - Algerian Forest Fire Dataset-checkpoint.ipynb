###### Finding inter quantile range:

```python
q1 = np.quantile(num_df['Temperature'],0.25)
q3 = np.quantile(num_df['Temperature'],0.75)
IQR = q3 - q1
print(f"q1 = {q1}, q3 = {q3}, IQR = {IQR}")
```

##### Finding outliers-free data using 5 number summary:

```python
lower_range = q1 - 1.5 * IQR
upper_range = q3 + 1.5 * IQR
outlier_free_list = [x for x in num_df['Temperature'] if ((x > lower_range) & (x < upper_range))]
```

```python
len(num_df['Temperature']) - len(outlier_free_list)
```

so, 2 outliers should be removed from the data (num_df['Temperature'])
