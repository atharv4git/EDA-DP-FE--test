```python
import pandas as pd
import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('nooutliers/retail_data_1.csv')
data.head()
```

```python
data.drop(columns='Unnamed: 0',axis=1,inplace=True)
```

I'll scale this dataset using only min max scaling for this time

```python
from sklearn.preprocessing import minmax_scale
minmax = minmax_scale(data.iloc[:,1:],feature_range=(0, 1))
data_minmax = pd.DataFrame(minmax,columns=data.iloc[:,1:].columns)
data_minmax['No'] = data['No']
data_minmax.head()
```

```python
import os
os.makedirs('scaled data')
data_minmax.to_csv('scaled data/real_estate_scaled_data.csv')
```
