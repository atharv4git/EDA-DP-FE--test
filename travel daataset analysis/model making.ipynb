```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('final csv/num_cat_combined1.csv')
data.head()
```

```python
data.drop(columns='Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.columns
```

# Splitting the data into train and test dataset

```python
from sklearn.model_selection import train_test_split
x = ['Age', 'DurationOfPitch', 'NumberOfFollowups','NumberOfTrips', 'NumberOfChildrenVisiting', 'MonthlyIncome','TypeofContact_Self Enquiry', 'TypeofContact_Company Invited', 'MaritalStatus_Single', 'MaritalStatus_Divorced','MaritalStatus_Married', 'MaritalStatus_Unmarried','Designation_Manager', 'Designation_Executive','Designation_Senior Manager', 'Designation_AVP', 'Designation_VP']
y = ['Occupation_Salaried', 'Occupation_Free Lancer','Occupation_Small Business', 'Occupation_Large Business']
feat_x = data[x]
feat_y = data[y]
x_train, x_test, y_train, y_test = train_test_split(feat_x, feat_y, test_size = 0.03)
```

# Classification using various models
## 1. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 1000)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

## 2. Dicision Tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)  
classifier.fit(x_train, y_train)
y_pred= classifier.predict(x_test)
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

## 3. K-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier= KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2 )  
classifier.fit(x_train, y_train)
y_pred= classifier.predict(x_test)
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

## 4. Naive Bayes

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(x_train, y_train).predict(x_test)
print("Number of mislabeled points out of a total %d points : %d"%(x_test.shape[0], (y_test != y_pred).sum()))
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```
