# Bifurcating data in numerical and catagorical data

```python
num_col=[fea for fea in data.columns if data[fea].dtype != 'O']
cat_col=[fea for fea in data.columns if data[fea].dtype == 'O']
```

```python
num_col,cat_col
```

```python
num_df = data[num_col]
cat_df = data[cat_col]
```

```python
cat_df
```

```python
num_df
```

# creating unique value dataframes for ease of analysis

```python
list1 = list()
for i in num_df.columns:
    list1.append(num_df[i].unique())

list2 = num_df.columns
unique_values_per_col_num = {
    "col": list2,
    "unique values": list1
}
unique_values_per_col_num_df = pd.DataFrame(unique_values_per_col_num)
```

```python
list1 = list()
for i in cat_df.columns:
    list1.append(cat_df[i].unique())

list2 = cat_df.columns
unique_values_per_col_cat = {
    "col": list2,
    "unique values": list1
}
unique_values_per_col_cat_df = pd.DataFrame(unique_values_per_col_cat)
```
