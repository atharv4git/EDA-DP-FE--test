```python
data.columns
```

```python
upper_list = []
lower_list = []
def removing_outliers(col):
    import sklearn
    print('for col:',col)
    # IQR
    Q1 = np.percentile(data[col], 25,interpolation = 'midpoint')

    Q3 = np.percentile(data[col], 75,interpolation = 'midpoint')
    IQR = Q3 - Q1

    print("Old Shape: ", data.shape)

    # Upper bound
    upper = np.where(data[col] >= (Q3+1.5*IQR))
    # Lower bound
    lower = np.where(data[col] <= (Q1-1.5*IQR))

    ''' Removing the Outliers '''
    upper_list.append(upper)
    lower_list.append(lower)

    print("New Shape: ", data.shape)
```

```python
for i in ['X3 distance to the nearest MRT station','X5 latitude','X6 longitude','Y house price of unit area']:
    removing_outliers(i)
```

```python
for i in lower_list:
    for j in i[0]:
        if j in data.index:
            data.drop(columns=data.columns.to_list(),index=j,axis=0,inplace=True)
```

```python
for i in lower_list:
    for j in i[0]:
        if j in data.index:
            data.drop(axis=0,inplace=True,index=j)
```

```python
for i in upper_list:
    for j in i[0]:
        if j in data.index:
            data.drop(axis=0,inplace=True,index=j)
```

```python
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(15, 12))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("Daily closing prices", fontsize=18, y=0.95)

for col, ax in zip(data.columns.to_list(), axs.ravel()):
    sns.boxplot(data[col],ax=ax)

    ax.set_title(col.upper())
    ax.set_xlabel("")

plt.show()
```
